```python
def train_val_test_split(X, Y, test_size=0.3, random_state=123):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, 
                                                        test_size=test_size, 
                                                        random_state=random_state)
   
    return X_train, X_test, Y_train, Y_test
```

```python
Y_condition1=Y[Y['condition']==1]
Y_condition2=Y[Y['condition']==2]
Y_condition3=Y[Y['condition']==3]
```

```python
XN100_condition1=X_N100[X_N100['condition']==1]
XN100_condition2=X_N100[X_N100['condition']==2]
XN100_condition3=X_N100[X_N100['condition']==3]
```

```python
XP200_condition1=X_P200[X_P200['condition']==1]
XP200_condition2=X_P200[X_P200['condition']==2]
XP200_condition3=X_P200[X_P200['condition']==3]
```

```python
XN100_condition1=XN100_condition1.drop('condition',axis=1)
XN100_condition2=XN100_condition2.drop('condition',axis=1)
XN100_condition3=XN100_condition3.drop('condition',axis=1)
Y_condiition1=Y_condition1.drop('condition', axis=1)
Y_condiition2=Y_condition2.drop('condition', axis=1)
Y_condiition3=Y_condition3.drop('condition', axis=1)
```

```python
XP200_condition1=XP200_condition1.drop('condition',axis=1)
XP200_condition2=XP200_condition2.drop('condition',axis=1)
XP200_condition3=XP200_condition3.drop('condition',axis=1)
```

```python
XN100_condition1_train, XN100_condition1_test, YN100_condition1_train, YN100_condition1_test = train_val_test_split(XN100_condition1, Y_condition1,
                                                                      
                                                                      test_size=0.3,
                                                                      random_state=123)
```

```python
XP200_condition1_train, XP200_condition1_test, YP200_condition1_train, YP200_condition1_test = train_val_test_split(XP200_condition1, Y_condition1,
                                                                      
                                                                      test_size=0.3,
                                                                      random_state=123)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
rkfold = RepeatedKFold(n_splits=7, n_repeats=3, random_state=42) 
param_grid = {'weights':['uniform', 'distance'],
'n_neighbors': [9, 11, 13, 15]}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=rkfold, return_train_score=True, scoring='accuracy', iid=False)
grid.fit(XN100_condition1_train, YN100_condition1_train)
grid.score(XN100_condition1_test, YN100_condition1_test), grid.best_params_, grid.best_score_
```

```python
rkfold = RepeatedKFold(n_splits=7, n_repeats=3, random_state=42) 
param_grid = {'weights':['uniform', 'distance'],
'n_neighbors': [9, 11, 13, 15]}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=rkfold, return_train_score=True, scoring='accuracy', iid=False)
grid.fit(XP200_condition1_train, YP200_condition1_train)
grid.score(XP200_condition1_test, YP200_condition1_test), grid.best_params_, grid.best_score_
```
